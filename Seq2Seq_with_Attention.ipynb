* Humans struggle to manually locate important information in large texts.

### **Extractive Summarization:**
* Identifies important sentences or phrases.
* Extracts these sentences to form the summary.
* Utilizes sentences directly from the original text.


### **Abstractive Summarization:**
* Generates new sentences from the original text.
* Creates summaries with sentences not present in the original content.
* Involves deep learning techniques for implementation.

## **Introduction to Sequence-to-Sequence (Seq2Seq) Modeling:**

* Seq2Seq models are used for sequential information problems.
* Applications include Sentiment Classification, Neural Machine Translation, and Named Entity Recognition.

### **Seq2Seq Model Architecture:**

A Seq2Seq model consists of two main components: Encoder and Decoder.

####**I)Encoder:**

**The Encoder** processes input data and captures its essential information.


Example: Given an input text "The weather is beautiful today," the encoder encodes its semantic meaning.

####**II)Decoder:**

**The Decoder** generates an output sequence based on the encoded information.

Example: Using the encoded meaning, the decoder creates a summary like "Today's weather is fantastic."

i) **Neural Machine Translation (NMT): Translates text from one language to another.**

II) **Named Entity Recognition (NER): Tags words in a sequence with corresponding labels.**

III)**Text Summarization as a Seq2Seq Problem:**
 Text summarization aims to condense a long text into a short summary.

##**Understanding the Encoder-Decoder Architecture:**

* Used for sequence-to-sequence (Seq2Seq) problems with differing input and output sequence lengths.
* Applied to text summarization, where long input text becomes a short output summary.


### **Encoder-Decoder Setup in Two Phases:**

* **Training Phase**

* **Inference Phase**


#### **Training Phase:**

* Set up encoder and decoder.
* Train model to predict target sequence with a one-timestep offset.
Encoder:
* Utilizes LSTM model to read input sequence word-by-word.
* Captures contextual information through hidden state (hi) and cell state (ci) of last timestep.


 **Decoder:**

* Also employs LSTM network.
* Predicts target sequence with one-timestep offset.
* Uses special tokens (<start> and <end>) for sequence initialization and completion.


#### **Inference Phase:**

* Applies model to new source sequences with unknown target sequences.
Inference Process:

* Encode entire input sequence and initialize decoder with encoder's internal states.
* Pass <start> token to decoder.
* Run decoder for one timestep, predicting next word's probability.
Select word with highest probability.
* Iterate until generating <end> token or hitting max sequence length.

##**Limitations of the Encoder-Decoder Architecture:**

**Challenges with Long Sequences:**

* The encoder-decoder architecture is effective for short input sequences.
* The encoder converts the entire input sequence into a fixed-length vector.
* This approach becomes problematic with long sequences.

**Difficulty in Handling Lengthy Inputs:**

* Encoding long sequences into a fixed-length vector is challenging.
* Memorizing extensive information into a limited vector can lead to loss of context.
* The architecture struggles to effectively capture the nuances of lengthy inputs.

**Performance Degradation with Longer Sentences:**

* As input sentence length increases, the basic encoder-decoder's performance deteriorates.
* The deterioration impacts the quality of predictions and overall output.

**Loss of Context and Information:**

* Compressing all essential information into a fixed-length vector can lead to the loss of critical context.
* This loss affects the model's ability to accurately understand and represent input sequences.

**Inadequate Handling of Long Sentences:**

* The architecture's inability to effectively manage long sentences hampers its practical utility.
* Longer sentences pose a challenge to the traditional encoder-decoder approach.

## **Attention Mechanism:**

![GIF Description](https://miro.medium.com/v2/resize:fit:1100/1*7Jbg-m9UNKXw-Mw_o-9BJQ.gif)


* The attention mechanism addresses how much attention is required for each word in the input sequence when generating a word at a specific timestep.

* This concept focuses on increasing the significance of certain portions of the source sequence to produce the desired target sequence.

##**Understanding through an Example:**
* Source sequence: "Which sport do you like the most?"
* Target sequence: "I love cricket"

* Connection between target words and source words, e.g., 'I' corresponds to 'you', 'love' to 'like'.

* Instead of uniformly considering all source words, attention emphasizes relevant parts for generating the target.


## **Basic Idea:**

Attention mechanism enhances important source sequence segments that contribute to generating the target sequence.
Classes of Attention Mechanism:

### **Global Attention:**

* Considers all source positions and hidden states of the encoder.
* Derives attended context vector from the complete source sequence.

### **Local Attention:**

* Focuses on a subset of source positions and corresponding hidden states.
* Attended context vector comes from selected encoder hidden states.

```python
input_text = list()
target_text = list()
input_words = list()
target_words = list()
stop_words = nltk.corpus.stopwords.words('english')
stemmer = nltk.PorterStemmer()
contractions = contraction_mapping
```

```python
import bs4
bs4.__version__
```

```python
def clean(text, src):
    # remove html tags
    texts = BeautifulSoup(text, "lxml").text
    
    # contractions
    words = [contractions[w] if w in contractions else w for w in nltk.word_tokenize(text.lower())]
    
    if src == "input":
        words = [stemmer.stem(w) for w in words if w not in stop_words]
    else:
        words = [w for w in words if w not in stop_words]
    
    return words
```

```python
for in_txt, tr_txt in zip(input_data, target_data):
    in_words = clean(in_txt, 'input')
    input_text += [" ".join(in_words)]
    input_words += in_words
    # add <eos> and <sos> token
    tr_words = clean("sos " + tr_txt + " eos", "target")
    target_text += [" ".join(tr_words)]
    target_words += tr_words
```

```python
import statistics
```

```python
input_words = sorted(list(set(input_words)))
target_words = sorted(list(set(target_words)))
num_in_words = len(in_words)
num_tr_words = len(tr_words)


max_in_len = statistics.mode([len(i) for i in input_text])
max_tr_len = statistics.mode([len(i) for i in target_text])
```

```python
print("number of input words : ",num_in_words)
print("number of target words : ",num_tr_words)
print("maximum input length : ",max_in_len)
print("maximum target length : ",max_tr_len)
```

```python
train_X, test_X, train_y, test_y = train_test_split(input_text, target_text, test_size=0.2)
```

```python
in_tokenizer = Tokenizer()
in_tokenizer.fit_on_texts(train_X)
tr_tokenizer = Tokenizer()
tr_tokenizer.fit_on_texts(train_y)

train_X = in_tokenizer.texts_to_sequences(train_X)
train_y = tr_tokenizer.texts_to_sequences(train_y)
```

```python
en_in_data = pad_sequences(train_X, maxlen=max_in_len, padding='post')
dec_data = pad_sequences(train_y, maxlen=max_tr_len, padding='post')

dec_in_data = dec_data[:,:-1]
dec_tr_data = dec_data.reshape(len(dec_data), max_tr_len, 1)[:, 1:]
```

```python
# Model Building
```

```python
K.clear_session()

latent_dim = 500
en_inputs = Input(shape=(max_in_len,))
en_embedding = Embedding(num_in_words+1, latent_dim)(en_inputs)
```

```python
en_lstm1 = LSTM(latent_dim, return_state=True, return_sequences=True)
en_outputs1, state_h1, state_c1 = en_lstm1(en_embedding)

en_lstm2 = LSTM(latent_dim, return_state=True, return_sequences=True)
en_outputs2, state_h2, state_c2 = en_lstm2(en_outputs1)

en_lstm3 = LSTM(latent_dim, return_state=True, return_sequences=True)
en_outputs3, state_h3, state_c3 = en_lstm3(en_outputs2)

# context vectors
en_states = [state_h3, state_c3]
```

```python
dec_inputs = Input(shape=(None,))
dec_emb_layer = Embedding(num_tr_words+1, latent_dim)
dec_embedding = dec_emb_layer(dec_inputs)


dec_lstm = LSTM(latent_dim, return_sequences=True, return_state=True)
dec_outputs, *_ = dec_lstm(dec_embedding, initial_state=en_states)
```

```python
attention = Attention()
attn_out = attention([dec_outputs, en_outputs3])

merge = Concatenate(axis=-1, name='concat_layer1')([dec_outputs, attn_out])

dec_dense = Dense(num_tr_words+1, activation='softmax')
dec_outputs = dec_dense(merge)
```

```python
model = Model([en_inputs, dec_inputs], dec_outputs)
model.summary()
plot_model(model, show_shapes=True, show_layer_names=True)
```
